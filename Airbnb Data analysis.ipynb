```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\rajat\Documents\Projects\Airbnb Data analysis\compressed_data.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month':0, 'last review' : df['last review'].min()}, inplace= True)
```

```python
df.isna().sum()
```

```python
df.dropna(subset= ['NAME', 'host name'], inplace= True)
```

```python
df = df.drop(columns = ['license','house_rules'], errors= 'ignore')
```

```python
df.info()
```

```python
df['price'] = df['price'].replace('[\$,]' , '', regex= True).astype(float)
```

```python
df['service fee'] = df['service fee'].replace('[\$,]' , '', regex= True).astype(float)
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace= True)
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,5))
sns.histplot(x= 'price', bins= 50,data = df, kde= True, color= 'green')
plt.title('Distribution of Listing Price')
plt.xlabel('Price $')
plt.ylabel('Frequency')
plt.show()
```

```python
df.info()
```

```python
df['room type'].value_counts()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data = df, x = 'room type', )
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(15,5))
sns.countplot(y = 'neighbourhood group', data = df , color= 'green' ,  order= df['neighbourhood group'].value_counts().index)
plt.title('Number of listing by neighbourhood groups')
plt.xlabel('Count')
plt.ylabel('Neighbourhood groups')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x = 'room type', y = 'price',hue='room type', data = df)
plt.title(r'Price vs Room Type')
plt.xlabel(r'Room Type')
plt.ylabel(r'Price $')
plt.show()
```

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(15,5))
reviews_over_time.plot(kind = 'line' ,color = 'red')
plt.title(r'Number of Reviews over time')
plt.xlabel(r'Date')
plt.ylabel(r'Reviews')
plt.show()
```
